```python
from bs4 import BeautifulSoup
import requests
import csv
import time
```

```python
# Data structure to store car details
Data_cars = []


for page in range(1200,1305):
    print(page)
    time.sleep(1)
    # URL of the website Tomobila
    url = f'https://www.tomobila.ma/achat-voiture-occasion/page/{page}'
    
    # Send the GET request
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all cars listed on the page
    cars = soup.find_all('div', {'class': 'listing-list-loop'})
    
    for car in cars:
        # Initialize a dictionary to store details of each car
        car_data = {
            'seller': '',
            'brand': '',
            'image': '',
            'price': '',
            'city': '',
            'origin': '',
            'fuel': '',
            'mileage': ''
        }
    
        # Extract the seller
        seller = car.find('div', {'class': 'meta-bottom'}).find('div', {'class': 'seller-title'})
        car_data['seller'] = seller.text.strip() if seller else 'Not available'
    
        # Extract the price
        price = car.find('div', {'class': 'meta-top'}).find('div', {'class': 'normal-price'})
        car_data['price'] = price.text.strip() if price else 'Not available'
    
        # Extract the brand
        brand = car.find('div', {'class': 'meta-top'}).find('div', {'class': 'labels'})
        car_data['brand'] = brand.text.strip() if brand else 'Not available'
    
        # Extract the city
        city = car.find('div', {'class': 'meta-middle'}).find('div', {'class': 'cities'})
        if city:
            city_value = city.find('div', {'class': 'value'})
            car_data['city'] = city_value.text.strip() if city_value else 'Not available'
        else:
            car_data['city'] = 'Not available'
    
        # Extract the fuel type
        fuel = car.find('div', {'class': 'meta-middle'}).find('div', {'class': 'fuel'})
        if fuel:
            fuel_value = fuel.find('div', {'class': 'value'})
            car_data['fuel'] = fuel_value.text.strip() if fuel_value else 'Not available'
        else:
            car_data['fuel'] = 'Not available'
    
        # Extract the origin
        origin = car.find('div', {'class': 'meta-middle'}).find('div', {'class': 'origin'})
        if origin:
            origin_value = origin.find('div', {'class': 'value'})
            car_data['origin'] = origin_value.text.strip() if origin_value else 'Not available'
        else:
            car_data['origin'] = 'Not available'
    
        # Extract the mileage
        mileage = car.find('div', {'class': 'meta-middle'}).find('div', {'class': 'mileage'})
        if mileage:
            mileage_value = mileage.find('div', {'class': 'value'})
            car_data['mileage'] = mileage_value.text.strip() if mileage_value else 'Not available'
        else:
            car_data['mileage'] = 'Not available'
    
        # Extract the image link
        image = car.find('div', {'class': 'stm-single-image'})
        if image:
            image_value = image.find('a', href=True)
            car_data['image'] = image_value['href'] if image_value else 'Not available'
        else:
            car_data['image'] = 'Not available'
    
        # Append the car data to the list
        Data_cars.append(car_data)
    
        # Print the extracted data
        print(car_data)
        print("-" * 40)
    
    # Print the final list of cars
    # print(Data_cars)
```

```python
# Open the CSV file for writing
with open('cars_tomobila.csv', 'a', newline='', encoding='utf-8') as csvfile:
    # Create a CSV writer object
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header = ['seller', 'brand', 'image', 'price', 'city', 'origin', 'fuel', 'mileage']
    spamwriter.writerow(header)

    # Write each car's details as a row
    for car in Data_cars:
        spamwriter.writerow([
            car['seller'],
            car['brand'],
            car['image'],
            car['price'],
            car['city'],
            car['origin'],
            car['fuel'],
            car['mileage']
        ])
```
